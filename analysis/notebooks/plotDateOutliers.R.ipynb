# Plot date from metadata and predicted date from Chronumental

```r
library(ggplot2)
library(cowplot)
library(forcats)
library(ggrepel)
library(dplyr)
library(reshape2)
library(patchwork)
```

```r
merged_actual_and_predicted_with_zscore <- read.csv("../outputFiles/chronumental/merged_actual_and_predicted_with_zscore.csv")
summary(merged_actual_and_predicted_with_zscore)
```

```r
merged_actual_and_predicted_with_zscore$date2 <- as.Date(merged_actual_and_predicted_with_zscore$date)
```

```r
merged_actual_and_predicted_with_zscore$predicted_date2 <- as.Date(merged_actual_and_predicted_with_zscore$predicted_date)
```

```r
ggplot(merged_actual_and_predicted_with_zscore, aes(x=date2, y=predicted_date2, color=outlier)) + 
theme_cowplot() +
geom_point(size = 0.5) +
labs(x = "Metadata collection date", y = "Time-tree predicted date")
```

```r
ggsave('../outputFiles/figures/metadata_date_vs_predicted_date.png', width=20, height=15, units = "cm")
```
