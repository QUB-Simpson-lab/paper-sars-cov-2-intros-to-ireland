```python
import pandas as pd
from tqdm.autonotebook import tqdm
tqdm.pandas()
import collections
from itertools import combinations
```

```python
GISAID_metadata = pd.read_csv('../datasets/GISAID/GISAID-hCoV-19-phylogeny-2022-05-20/metadata.csv')
GISAID_metadata = GISAID_metadata[(GISAID_metadata['virus_name'].str.startswith('hCoV-19/Northern Ireland/')) | (GISAID_metadata['virus_name'].str.startswith('hCoV-19/Ireland/'))].copy()
GISAID_metadata
```

```python
def add_type(cell):
    fields = cell.split('/')
    
    if fields[1] == 'Northern Ireland':
        ID_parts = fields[2].split('-')
        return ID_parts[0]
    
    if fields[1] == 'Ireland':
        return 'Ireland'
```

```python
GISAID_metadata['type'] = GISAID_metadata['virus_name'].apply(add_type)
GISAID_metadata
```

```python
clusters_timepoint_A = pd.read_csv('../outputFiles/clusters/clusters_timepoint_A.csv')
clusters_timepoint_B = pd.read_csv('../outputFiles/clusters/clusters_timepoint_B.csv')
clusters_timepoint_C = pd.read_csv('../outputFiles/clusters/clusters_timepoint_C.csv')
clusters_timepoint_D = pd.read_csv('../outputFiles/clusters/clusters_timepoint_D.csv')
clusters_timepoint_E = pd.read_csv('../outputFiles/clusters/clusters_timepoint_E.csv')
clusters_timepoint_F = pd.read_csv('../outputFiles/clusters/clusters_timepoint_F.csv')

clusters = pd.concat([clusters_timepoint_A,
           clusters_timepoint_B,
           clusters_timepoint_C,
           clusters_timepoint_D,
           clusters_timepoint_E,
           clusters_timepoint_F])

clusters['cluster_timepoint'] = clusters['timepoint'] + '_' + clusters['cluster'].astype(str) 
clusters
```

```python
clusters = pd.merge(clusters, GISAID_metadata, how='inner', left_on='tip', right_on='accession_id')
clusters
```

```python
clusters['irish_in_cluster'] = 'N'

for index, row in tqdm(clusters.iterrows()):
    if len(clusters[(clusters['cluster_timepoint'] == row['cluster_timepoint']) & (clusters['type'] == 'Ireland')]) > 0:
        clusters.loc[index,'irish_in_cluster'] = 'Y'
```

```python
clusters[(clusters['type'] == 'NORT') & (clusters['irish_in_cluster'] == 'Y') & (clusters['NIRE_in_cluster'] == 'Y')]
```

```python
ID_type_cluster_irish_counts = clusters.groupby(['type','irish_in_cluster']).count().reset_index()[['type','irish_in_cluster','tip']]
ID_type_cluster_irish_counts = ID_type_cluster_irish_counts[ID_type_cluster_irish_counts['type'] != 'Ireland']
ID_type_cluster_irish_counts.to_csv('../outputFiles/clusters/NorthernIrelandIDTypeClusterSharedWithRoISampleCount.csv', index=False)
ID_type_cluster_irish_counts
```
