```python
import pandas as pd
```

```python
NI_cases = pd.read_csv('../datasets/CovidDashboardData/UK/Cases_NI_data_2022-Jun-22.csv')
NI_cases['date']= pd.to_datetime(NI_cases['date'])
NI_cases.sort_values(by=['date'], ascending=True, inplace=True)
NI_cases.index = pd.DatetimeIndex(NI_cases['date'])
NI_cases = NI_cases.reindex(pd.date_range('2020-02-26', '2022-05-19'))
NI_cases = NI_cases.reset_index()
NI_cases['newCasesBySpecimenDate'].fillna(0, inplace=True)
NI_cases['7_day_rolling_cases'] = NI_cases['newCasesBySpecimenDate'].rolling(window=7,min_periods=0).mean()
# Population estimate from 2019 Mid-year Population Estimates for Northern Ireland - https://www.nisra.gov.uk/sites/nisra.gov.uk/files/publications/MYE19-Bulletin.pdf
NI_cases['7_day_cases_per_100k'] = NI_cases['7_day_rolling_cases'] / 1893700 * 100000
NI_cases
```

```python
NI_hospitalisations = pd.read_csv('../datasets/CovidDashboardData/UK/HospitalAdmissions_2022-Jun-22.csv')
NI_hospitalisations['date']= pd.to_datetime(NI_hospitalisations['date'])
NI_hospitalisations.sort_values(by=['date'], ascending=True, inplace=True)
NI_hospitalisations['7_day_hospitalisations_rolling'] = NI_hospitalisations['newAdmissions'].rolling(window=7,min_periods=0).mean()
NI_hospitalisations['7_day_hospitalisations_per_100k'] = NI_hospitalisations['7_day_hospitalisations_rolling'] / 1893700 * 100000
NI_hospitalisations
```

```python
NI_deaths = pd.read_csv('../datasets/CovidDashboardData/UK/DeathsByDateOfDeath_NI_data_2022-Jun-22.csv')
NI_deaths['date']= pd.to_datetime(NI_deaths['date'])
NI_deaths.sort_values(by=['date'], ascending=True, inplace=True)
NI_deaths['7_day_deaths_rolling'] = NI_deaths['newDeaths28DaysByDeathDate'].rolling(window=7,min_periods=0).mean()
NI_deaths['7_day_deaths_per_100k'] = NI_deaths['7_day_deaths_rolling'] / 1893700 * 100000
NI_deaths
```

```python
NI = pd.merge(
    NI_cases[['index','areaName', 'date','newCasesBySpecimenDate','7_day_rolling_cases','7_day_cases_per_100k']],
    NI_hospitalisations[['date','newAdmissions','7_day_hospitalisations_rolling','7_day_hospitalisations_per_100k']],
    how="left",
    on=['date']
)
NI

NI = pd.merge(
    NI,
    NI_deaths[['date','newDeaths28DaysByDeathDate','7_day_deaths_rolling','7_day_deaths_per_100k']],
    how="left",
    on=['date']
)
NI
```

```python
ROI_data = pd.read_csv('../datasets/CovidDashboardData/Ireland/COVID-19_HPSC_Detailed_Statistics_Profile.csv')
ROI_data['date'] = pd.to_datetime(pd.to_datetime(ROI_data['Date']).dt.date)
ROI_data.sort_values(by=['date'], ascending=True, inplace=True)
ROI_data.index = pd.DatetimeIndex(ROI_data['date'])
ROI_data = ROI_data.reindex(pd.date_range('2020-02-29', '2022-06-22'))
ROI_data = ROI_data.reset_index()
ROI_data['ConfirmedCovidCases'].fillna(0, inplace=True)

ROI_data['7_day_rolling_cases'] = ROI_data['ConfirmedCovidCases'].rolling(window=7,min_periods=0).mean()

ROI_data['newAdmissions'] = ROI_data['HospitalisedCovidCases'].diff()
ROI_data['newAdmissions'].fillna(0, inplace=True)
ROI_data['7_day_hospitalisations_rolling'] = ROI_data['newAdmissions'].rolling(window=7,min_periods=0).mean()

ROI_data['DeathsToday_DOD'].fillna(0, inplace=True)
ROI_data['7_day_deaths_rolling'] = ROI_data['DeathsToday_DOD'].rolling(window=7,min_periods=0).mean()

ROI_data['7_day_cases_per_100k'] = ROI_data['7_day_rolling_cases'] / 4761865 * 100000
ROI_data['7_day_hospitalisations_per_100k'] = ROI_data['7_day_hospitalisations_rolling'] / 4761865 * 100000
ROI_data['7_day_deaths_per_100k'] = ROI_data['7_day_deaths_rolling'] / 4761865 * 100000
ROI_data['areaName'] = 'Republic of Ireland'
ROI_data
```

```python
irish_case_data = pd.concat([
    NI[['index',
        'date',
        '7_day_cases_per_100k',
        '7_day_hospitalisations_per_100k',
        '7_day_deaths_per_100k',
        'areaName']],
    ROI_data[['index',
              'date',
              '7_day_cases_per_100k',
              '7_day_hospitalisations_per_100k',
              '7_day_deaths_per_100k',
              'areaName']]])
irish_case_data
```

```python
irish_case_data.to_csv('../outputFiles/irish_case_data.csv', index=False)
```

```python
full_GISAID_metadata = pd.read_csv('../datasets/GISAID/metadata_tsv_2022_06_04/metadata.tsv', sep='\t')
full_GISAID_metadata.head()
```

```python
def simplified_country(row):
    location = ''
    if row['Location'].startswith('Europe / United Kingdom / Northern Ireland'):
        location = 'Northern Ireland'
    elif row['Location'].startswith('Europe / Ireland'):
        location = 'Republic of Ireland'
    else:
        location = 'Other'
    return location

full_GISAID_metadata['Simple country'] = full_GISAID_metadata.apply(simplified_country, axis = 1)
```

```python
irish_GISAID_metadata = full_GISAID_metadata[full_GISAID_metadata['Simple country'].isin(['Northern Ireland', 'Republic of Ireland'])].copy()
irish_GISAID_metadata.groupby('Simple country').count()
```

```python
irish_GISAID_metadata.to_csv('../datasets/GISAID/metadata_tsv_2022_06_04/metadata.RoIAndNI.tsv', sep='\t', index=False)
```
