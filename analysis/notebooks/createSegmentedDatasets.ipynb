#### Convert date and extract country information

````python
full_GISAID_metadata['Collection date 2'] = pd.to_datetime(full_GISAID_metadata['Collection date'])
````

````python
def extract_country(row):
    location = ''
    if row['Location'].startswith('Europe / United Kingdom / Northern Ireland'):
        location = 'Northern Ireland'
    elif row['Location'].startswith('Europe / United Kingdom / Scotland'):
        location = 'Scotland'
    elif row['Location'].startswith('Europe / United Kingdom / England'):
        location = 'England'
    elif row['Location'].startswith('Europe / United Kingdom / Wales'):
        location = 'Wales'
    elif row['Location'].startswith('Europe / Ireland'):
        location = 'Republic of Ireland'
    else:
        location = ' / '.join(row['Location'].split(' / ')[0:2])
    return location

def simplified_country(row):
    location = ''
    if row['Location'].startswith('Europe / United Kingdom / Northern Ireland') or row['Location'].startswith('Europe / Ireland'):
        location = 'Ireland'
    else:
        location = 'Other'
    return location

full_GISAID_metadata['Country'] = full_GISAID_metadata.apply(extract_country, axis = 1)
full_GISAID_metadata['Simple country'] = full_GISAID_metadata.apply(simplified_country, axis = 1)
````

````python
chronumental_dates = pd.read_csv('../outputFiles/chronumental/chronumental_dates_date_file.tsv.tsv', sep="\t")
chronumental_dates
````

````python
merged_dates = pd.merge(date_file,chronumental_dates,on="strain")
merged_dates
````

````python
merged_dates['date'] = pd.to_datetime(merged_dates['date'])
merged_dates['predicted_date'] = pd.to_datetime(merged_dates['predicted_date'])
````

### Calculate difference between metadata date and predicted date

````python
merged_dates['date_diff'] = merged_dates['date'] - merged_dates['predicted_date']
````

````python
merged_dates
````

````python
merged_dates.describe()
````

### Calculate Z-score of date difference

````python
merged_dates['zscore'] = (merged_dates['date_diff'] - merged_dates['date_diff'].mean()) / merged_dates['date_diff'].std()
merged_dates
````

````python
merged_dates[(merged_dates['zscore'] > 3) | (merged_dates['zscore'] < -3)]
````

### Get list of date outliers by Z-score

````python
date_outliers = merged_dates[(merged_dates['zscore'] > 3) | (merged_dates['zscore'] < -3)]['strain']
date_outliers
````

````python
merged_dates['outlier'] = False
merged_dates.loc[(merged_dates['zscore'] > 3) | (merged_dates['zscore'] < -3), 'outlier'] = True
````

````python
merged_dates.groupby('outlier').count()
````

````python
merged_dates.groupby('outlier').describe()
````

````python
merged_dates[(merged_dates['zscore'] < -2.999) & (merged_dates['zscore'] > -3.001)]
````

````python
merged_dates[(merged_dates['zscore'] > 2.999) & (merged_dates['zscore'] < 3.001)]
````
